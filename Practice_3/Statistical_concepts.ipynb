# 3. Three sigma rule (66 / 95 / 99.7)

```python
# mean - 3 * std  mean + 3 * std
```

```python
mean = df.salary_in_usd.mean()
std = df.salary_in_usd.std()


# 3 sigma rule
upper_bound = mean + 3 * std
lower_bound = mean - 3 * std # 99.7%

# 2 sigma rule
upper_bound_1 = mean + 2 * std # 95%
lower_bound_1 = mean - 2 * std

# 1 sigma rule
upper_bound_2 = mean + 1 * std # 66.7%
lower_bound_2 = mean - 1 * std
```

```python
sns.histplot(df.salary_in_usd)

plt.axvline(upper_bound, color='r', label='$3 \sigma$', ls='--')
plt.axvline(lower_bound, color='r', ls='--')

plt.axvline(upper_bound_1, color='y', ls='--', label='$2 \sigma$')
plt.axvline(lower_bound_1, color='y', ls='--')

plt.axvline(upper_bound_2, color='c', ls='--', label='$1 \sigma$')
plt.axvline(lower_bound_2, color='c', ls='--')

plt.axvline(mean, color='purple', label='mean')
plt.legend()
plt.show()
```

# 4. IQR rule
IQR = inter quantile range

IQR = Q75 - Q25

```python
Q75  = np.percentile(df.salary_in_usd, 75)
Q25  = np.percentile(df.salary_in_usd, 25)
Q50  = np.percentile(df.salary_in_usd, 50)

IQR = Q75 - Q25

lower_bound = Q25 - 1.5 * IQR
upper_bound = Q75 + 1.5 * IQR
```

```python
sns.histplot(df.salary_in_usd)
plt.axvline(upper_bound, color='c', label='3Q + 1.5IQR', ls='--')
plt.axvline(lower_bound, color='g', label='1Q - 1.5IQR', ls='--')
plt.title('IQR rule')
plt.axvline(Q50, color='r', label='Median')
plt.legend()
plt.show()
```

# 5. Rule/Law of big numbers

```python
n_experiments = 1000
experiments = np.zeros(n_experiments)
for i in range(1, n_experiments):
    experiments[i] = np.random.randint(2, size=i).mean()
```

```python
plt.plot(experiments, linewidth=.5)
plt.title('Law of big numbers')
plt.xlabel('N experiments')
plt.axhline(0.5, label='Expected value', color='r')
plt.ylabel('Probability of tail')
plt.legend()
plt.grid(alpha=.3)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 4), dpi=150)
ax1.plot(experiments, linewidth=.7)
ax1.set_title('Tail probability over the experiments')
ax2.set_title('Distribution of tail probabilities')

ax1.set_xlabel('N experiments')
ax1.axhline(0.5, label='Expected value', color='r')
ax2.axvline(0.5, label='Expected value', color='r')
ax1.set_ylabel('Probability of tail')

sns.histplot(experiments, ax=ax2)
# ax1.legend()
# ax1.grid(alpha=.3)
plt.show()
```

# 6. Central Limit Theorem (CLT)

```python
gamma_distr = np.random.gamma(shape=2, size=100_000)

sns.histplot(gamma_distr, kde=True)
plt.title('Gamma dist')
plt.show()
```

```python
subsample = np.random.choice(gamma_distr, size=1000)
mean_subsample = np.mean(subsample)

sns.histplot(subsample, kde=True)
plt.title(f'Sample Mean = {mean_subsample:.3f}')
plt.show()
```

```python
subsample = np.random.choice(gamma_distr, size=1000)
mean_subsample = np.mean(subsample)

sns.histplot(subsample, kde=True)
plt.title(f'Sample Mean = {mean_subsample:.3f}')
plt.show()
```

```python
n_experiments = 10_000
experiments_means = np.zeros(n_experiments)
for i in range(n_experiments):
    subsample = np.random.choice(gamma_distr, size=1000)
    mean_subsample = np.mean(subsample)
    experiments_means[i] = mean_subsample

sns.histplot(experiments_means, kde=True)
plt.show()
```

```python
def central_limiting_theorem(func, n_experiments = 10_000, **kwargs):
    distr = func(**kwargs, size=100_000, )
    
    experiments_means = np.zeros(n_experiments)
    for i in range(n_experiments):
        subsample = np.random.choice(distr, size=1000)
        mean_subsample = np.mean(subsample)
        experiments_means[i] = mean_subsample
        
    
    fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(12, 4), dpi=150)
    sns.histplot(distr, ax=ax1, kde=True)
    ax1.set_title('Initial distribution')

    subsample = np.random.choice(distr, size=1000)
    mean_subsample = np.mean(subsample)

    sns.histplot(subsample, ax=ax2, kde=True)
    ax2.set_title(f'Sample Distr. with a Mean={mean_subsample:.3f}')


    sns.histplot(experiments_means, kde=True, ax=ax3)
    ax3.set_title(f'Distr. of {n_experiments} subsamples mean')
        
    plt.show()
```

## 6.1 CLT with initial Normal distribution

```python
central_limiting_theorem(np.random.normal)
```

## 6.2 CLT with initial uniform distribution

```python
central_limiting_theorem(np.random.uniform)
```

## 6.3 CLT with initial gamma distribution

```python
central_limiting_theorem(np.random.gamma, shape = 3)
```

## 6.4 CLT with initial exponential distribution

```python
central_limiting_theorem(np.random.exponential)
```
